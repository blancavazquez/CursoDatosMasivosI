## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/data_fraudes.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create una tabla temporal llamada "fraudes_data_csv"
temp_table_name = "fraudes_data_csv"
df.createOrReplaceTempView(temp_table_name)
```

```
#Creamos un dataFrame
df = spark.sql("select step, type, amount, nameOrig, oldbalanceOrg, newbalanceOrig, nameDest, oldbalanceDest, newbalanceDest from fraudes_data_csv")
```

```
%sql 
/* Esta es una celda para ejecutar consultas SQL sobre la tabla creada "fraudes_data_csv" */
/* Visualizar cuntas transacciones tiene la tabla agrupada por tipo de transaccin*/
select type, count(*) from `fraudes_data_csv` group by type
```

```
%sql
/* Visualizar la cantidad de dinero transferido por cada tipo de transaccin */
select type, sum(amount) from `fraudes_data_csv` group by type
```

```
from pyspark.sql import functions as F
#Generacin patrones para detectar fraudes (modelo basado en reglas)
#La funcin de spack withcolumn se usa para actualizar el valor de una columna existente o crear una nueva columna en un dataset
df = df.withColumn("label", #creamos una nueva columna "label": "1" es fraude, "0" no es fraude 
                   F.when(
                     ((df.oldbalanceOrg <= 56900) & (df.type == "TRANSFER") & (df.newbalanceDest <= 105)) | 
                     ((df.oldbalanceOrg > 56900) & (df.newbalanceOrig <= 12)) | 
                     ((df.oldbalanceOrg > 56900) & (df.newbalanceOrig > 12) & (df.amount > 1160000)
                     ), 1
                     ).otherwise(0))
```

```
df.show(20)
```

```
#Aadimos una nueva columna
df = df.withColumn("Limit_max",
                   F.when(
                     ((df.amount >= 1000) & (df.type == "TRANSFER") 
                     ), 1
                     ).otherwise(0))
```

```
df.show(20)
```

```
#Renombramos una columna "nameDest" por "destinatario"
df = df.withColumnRenamed("nameDest","destinatario")
```

```
#Renombramos una columna "label" por "label_fraude"
df = df.withColumnRenamed("label","label_fraude")
```

```
#Eliminamos una columna
df = df.drop("destinatario")
```

```
#Creamos una tabla temporal con los datos etiquetados
temp_table_name = "fraudes_labeled_csv"
df.createOrReplaceTempView(temp_table_name)
```

```
%sql 
select label_fraude, count(*) as `Transacciones`, 
       sum(amount) as `Cantidad total` 
from   `fraudes_labeled_csv` 
group by label_fraude
```

```
#Iniciamos la construccin del modelo de aprendizaje
#Paso 1: separamos los datos en entrenamiento y validacin
(train, val) = df.randomSplit([0.8, 0.2], seed=12345)
print("Total rows: %s, Training rows: %s, Test rows: %s" % (df.count(), train.count(), val.count()))
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

#Paso 2: convertirmos la variable categrica "type" a una variable numrica (usando .StringIndexer)
indexer = StringIndexer(inputCol = "type", outputCol = "typeIndexed")

#Paso 3: seleccionamos las columnas que utilizaremos para el modelo
# VectorAssembler es un transformer que combina dada una lista de columnas hacias un vector de una sola columna
va = VectorAssembler(inputCols = ["typeIndexed", "amount", "oldbalanceOrg", "newbalanceOrig", "oldbalanceDest", "newbalanceDest"], outputCol = "features")

#Paso 4: declaramos el mtodo a entrenar
dt = DecisionTreeClassifier(labelCol = "label_fraude", featuresCol = "features", seed = 54321, maxDepth = 5)

#Paso 5: creamos un pipeline
pipeline = Pipeline(stages=[indexer, va, dt])

#Paso 6: entrenamos el modelo
dt_model = pipeline.fit(train)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator

#Mtricas para evaluar el rendimiento del modelo
evaluatorPR = BinaryClassificationEvaluator(labelCol = "label_fraude", 
                                            rawPredictionCol = "prediction", metricName = "areaUnderPR")
evaluatorAUC = BinaryClassificationEvaluator(labelCol = "label_fraude", 
                                             rawPredictionCol = "prediction", metricName = "areaUnderROC")
```

```
#Vamos a buscar los mejores parmeteros para entrenar al modelo

#Importamos MLflow: librera para aprendizaje mquina
import mlflow
import mlflow.spark
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

# Construimos un gridSearch para evaluar diferentes parmetros
paramGrid = ParamGridBuilder() \
.addGrid(dt.maxDepth, [5, 10, 15]) \
.addGrid(dt.maxBins, [10,20,30]) \
.build()

# Tratamos el problema de desbalanceo de clases
crossval = CrossValidator(estimator = dt,   #Recordemos que el estimador ya se defini previamente dt = DecisionTreeClassifier(labelCol = "label_fraude", featuresCol = "features", seed = 54321, maxDepth = 5)
                          estimatorParamMaps = paramGrid,
                          evaluator = evaluatorPR, #mtrica usada para clases desbalanceadas
                          numFolds = 3)  
# Construimos un pipeline
pipelineCV = Pipeline(stages=[indexer, #indexer = StringIndexer(inputCol = "type", outputCol = "typeIndexed")
                              va,  #va = VectorAssembler(inputCols = ["typeIndexed", "amount", "oldbalanceOrg", "newbalanceOrig", "oldbalanceDest", "newbalanceDest"], outputCol = "features")
                              crossval])
```

```
# Entrenamos nuevamente el modelo usando el pipeline, parameter grid, y obtenemos la mtrica de evaluacin
cvModel_u = pipelineCV.fit(train) #El resultado es la identificacin de los mejores parmetros
```

```
#Hacemos predicciones usando el conjunto de entrenamiento
train_pred = cvModel_u.transform(train)
#Hacemos predicciones usando el conjunto de validacin
val_pred = cvModel_u.transform(val)

#Evaluamos el rendimiento (train)
pr_train = evaluatorPR.evaluate(train_pred)
auc_train = evaluatorAUC.evaluate(train_pred)

#Evaluamos el rendimiento (val)
pr_val = evaluatorPR.evaluate(val_pred)
auc_val = evaluatorAUC.evaluate(val_pred)
```

```
#Impresin de los resultados:
print("AUC train:", auc_train)
print("AUC test:", auc_val)
```
